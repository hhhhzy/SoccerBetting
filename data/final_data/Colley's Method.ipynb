### Let us start with Using Goals as winner indicator

Adjust for home field advantage

```python
df = df_copy.copy()
```

```python
df = df[['game_id', 'League', 'Season', 'Date', 'Team_Home', 'Team_Away', 'Goals_Home', 'Goals_Away']]
```

```python
# df['Goals_Home'] = df['Goals_Home'] / 1.5
df['Goals_Home'] = df['Goals_Home'] 
# df['Goals_Away'] = df['Goals_Away'] / 1.1
df['Goals_Away'] = df['Goals_Away'] 
```

```python
df['Win_Home'] = 1 * (df['Goals_Home'] > df['Goals_Away'])
df['Win_Away'] = 1 * (df['Goals_Home'] < df['Goals_Away'])

df['Loss_Home'] = 1 * (df['Goals_Home'] < df['Goals_Away'])
df['Loss_Away'] = 1 * (df['Goals_Home'] > df['Goals_Away'])

df['Draw_Home'] = 1 * (df['Goals_Home'] == df['Goals_Away'])
df['Draw_Away'] = 1 * (df['Goals_Home'] == df['Goals_Away'])
# df['Draw'] = 1 * (df['Goals_Home'] == df['Goals_Away'])
```

```python
df
```

```python
df_long = pd.wide_to_long(df, ["Goals", 'Team', 'Win', 'Loss', 'Draw'], i="game_id", j="isHome", sep='_', suffix='(!?Home|Away)')
df_long = df_long.reset_index()
df_long = df_long.sort_values(['Date', 'game_id'])
df_long
```

```python
df_long["TotalGamesPlayed"] = df_long.groupby(['Season', 'League', 'Team'])['Win'].transform(lambda x: pd.Series(range(1, len(x) + 1)))
df_long["TotalWins"] = df_long.groupby(['Season', 'League', 'Team'])['Win'].transform(lambda x: x.cumsum())
df_long["TotalDraws"] = df_long.groupby(['Season', 'League', 'Team'])['Draw'].transform(lambda x: x.cumsum())
df_long["TotalLosses"] = df_long.groupby(['Season', 'League', 'Team'])['Loss'].transform(lambda x: x.cumsum())
```

```python
home_df = df_long[df_long['isHome'] == 'Home'].drop('isHome', 1)
away_df = df_long[df_long['isHome'] == 'Away'].drop('isHome', 1)
home_df.head(1)
```

```python
new_df = pd.merge(home_df, away_df, how = 'inner', on = ['Season', 'League', 'game_id', 'Date'], suffixes=('_Home', '_Away'))
new_df.head()
```

```python
new_df
```

```python
from itertools import combinations
from collections import defaultdict
import statsmodels.api as sm

subset_dfs = []

for season in new_df['Season'].unique():
    for league in new_df['League'].unique():
        subset_df = new_df[(new_df['Season'] == season) & (new_df['League'] == league)].sort_values(['Date','game_id'])
        
        teams = list(subset_df['Team_Home'].unique())
        C = np.eye(len(teams), len(teams)) * 2
        b = np.ones((len(teams), 1))
        
        same_teams_hash = {}
        subset_df['ColleyRating_Home'] = 0
        subset_df['ColleyRating_Away'] = 0
                
        for i, row in subset_df.iterrows():
            home_team = row['Team_Home']
            away_team = row['Team_Away']
            
            ih = teams.index(home_team)
            ia = teams.index(away_team)
            
            b[ih] = 1 + (0.5 * (row['TotalWins_Home'] - row['TotalLosses_Home']))
            b[ia] = 1 + (0.5 * (row['TotalWins_Away'] - row['TotalLosses_Away']))       
            
            C[ih, ih] = 2 + row['TotalGamesPlayed_Home']
            C[ia, ia] = 2 + row['TotalGamesPlayed_Away']
            
            if str((ia, ih)) in same_teams_hash or str((ih, ia)) in same_teams_hash:
                
                if str((ia, ih)) in same_teams_hash:
                    same_teams_hash[str((ia, ih))] += 1
                    C[ih, ia] = - same_teams_hash[str((ia, ih))]
                    C[ia, ih] = - same_teams_hash[str((ia, ih))]
                else:
                    same_teams_hash[str((ih, ia))] += 1
                    C[ih, ia] = - same_teams_hash[str((ih, ia))]
                    C[ia, ih] = - same_teams_hash[str((ih, ia))]
            else:
                same_teams_hash[str((ih, ia))] = 1
                C[ih, ia] = -1
                C[ia, ih] = -1
                
            model = sm.OLS(b, C).fit()
            
            rating_dict = dict(zip(teams, model.params))
            
            subset_df.loc[i, 'ColleyRating_Home'] = rating_dict[home_team]
            subset_df.loc[i, 'ColleyRating_Away'] = rating_dict[away_team]
        
        subset_dfs.append(subset_df)
        
```

```python
colley_df = pd.concat(subset_dfs, ignore_index = True).sort_values(['Date', 'game_id'])
colley_df
```

```python
colley_df['ColleyRatingVar'] = colley_df['ColleyRating_Home'] - colley_df['ColleyRating_Away']
colley_df['Win_Home'] = 1 * ((colley_df['Goals_Home']) > (colley_df['Goals_Away']))
```

```python
from sklearn.metrics import brier_score_loss

import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
cols_for_pred = [
#     'FTEProb_Home',
    'ColleyRatingVar'
]

glm_str = "Win_Home ~ " + '+'.join(cols_for_pred)
print(glm_str)

result_glm = smf.glm(glm_str, 
             data = colley_df[colley_df['Season'].isin(['2017-2018', '2018-2019'])], 
             family = sm.families.Binomial()
).fit()
result_glm.summary()
```

```python
print("Validation Brier Score: ", brier_score_loss(colley_df[colley_df['Season'].isin(['2019-2020'])]['Win_Home'], result_glm.predict(colley_df[colley_df['Season'].isin(['2019-2020'])])))
```

```python
from sklearn.metrics import accuracy_score
print("Validation Brier Score: ", accuracy_score(colley_df[colley_df['Season'].isin(['2019-2020'])]['Win_Home'], 1 * (result_glm.predict(colley_df[colley_df['Season'].isin(['2019-2020'])]) > 0.5)))
```
