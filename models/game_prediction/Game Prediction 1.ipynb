```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/final_data/final_data.csv', parse_dates = ['Date'])
df = df.rename(columns = {
    'FBREFSave%_Home': 'FBREFSavePerc_Home',
    'FBREFSave%_Away': 'FBREFSavePerc_Away'
})
colley_df = pd.read_csv('../../data/final_data/colley/matches.csv', parse_dates = ['Date'])
```

```python
rename_cols = {}
for col in df.columns:
    if col.startswith('538'):
        rename_cols[col] = col.replace('538', 'FTE')
rename_cols
```

```python
df = df.rename(columns = rename_cols)
```

```python
df
```

```python
df.columns
```

```python
df_copy = df.copy()
```

```python
df['GameID'] = df.index
```

```python
drop_cols1 = []
for col in df.columns:
    if col.endswith('_Draw'):
        drop_cols1.append(col)
df = df.drop(drop_cols1, 1)
```

```python
df[['FTExg_Home', 'FBREFxg_Home', 'USxg_Home', 'FTExg_Away', 'FBREFxg_Away', 'USxg_Away']]
```

```python
from collections import defaultdict

col_names_with_home_or_away = defaultdict(lambda: 0)
for col in df.columns:
    if col.endswith('_Home') or col.endswith('_Away'):
        col_names_with_home_or_away[''.join(col.split('_')[:-1])] += 1

for col in col_names_with_home_or_away.keys():
    assert col_names_with_home_or_away[col] == 2, print(col)
    
df_long = pd.wide_to_long(
    df, 
    col_names_with_home_or_away, 
    i = ['GameID'], 
    j = 'isHome', 
    sep = '_', 
    suffix = r'\w+'
)
df_long = df_long.reset_index().sort_values(['Date','GameID'])

df_long
```

```python
df_long[['GameID', 'isHome', 'Date', 'Season', 'League', 'Team', 'Goals', 'FTEImportance']]
```

```python
df_long.columns
```

```python
df_long['Game#'] = df_long.groupby(['League', 'Team'])['FBREFORating'].transform(lambda x: pd.Series(range(0, len(x)), index=x.index))
```

```python
df_long['AvgFBREFORating'] = df_long.groupby(['League', 'Team'])['FBREFORating'].transform(lambda x: x.rolling(15, 1).mean().shift(1, fill_value=0))
df_long['AvgFBREFDRating'] = df_long.groupby(['League', 'Team'])['FBREFDRating'].transform(lambda x: x.rolling(15, 1).mean().shift(1, fill_value=0))

df_long['AvgFBREFORating2'] = df_long.groupby(['League', 'Team'])['FBREFORating2'].transform(lambda x: x.rolling(15, 1).mean().shift(1, fill_value=0))
df_long['AvgFBREFDRating2'] = df_long.groupby(['League', 'Team'])['FBREFDRating2'].transform(lambda x: x.rolling(15, 1).mean().shift(1, fill_value=0))

df_long['AvgFBREFSavePerc'] = df_long.groupby(['League', 'Team'])['FBREFSavePerc'].transform(lambda x: x.rolling(15, 1).mean().shift(1, fill_value=0))
# FBREFGCA_Home
df_long['AvgFBREFGCA'] = df_long.groupby(['League', 'Team'])['FBREFGCA'].transform(lambda x: x.rolling(15, 1).mean().shift(1, fill_value=0))
df_long['AvgFBREFSCA'] = df_long.groupby(['League', 'Team'])['FBREFSCA'].transform(lambda x: x.rolling(15, 1).mean().shift(1, fill_value=0))
```

```python
df_long = df_long.dropna()
```

```python
# df_long = df_long[df_long['Game#'] > 4]
```

```python
new_df_home = df_long[df_long['isHome'] == 'Home'].drop('isHome', 1)
new_df_away = df_long[df_long['isHome'] == 'Away'].drop('isHome', 1)

final_df = pd.merge(new_df_home, new_df_away, how='inner', on=['GameID', 'Date', 'Season', 'League'], suffixes=('_Home', '_Away'))
final_df.columns, len(final_df)
```

```python
final_df['Win_Home'] = 1 * (final_df['Goals_Home'] > final_df['Goals_Away'])
```

```python
final_df
```

```python
final_df.columns
```

```python
final_df = pd.merge(final_df, colley_df.drop(['Goals_Home', 'Goals_Away'], 1), how='inner', on=[
    'Season', 'League', 'Date', 'Team_Home', 'Team_Away'
])
```

```python
from sklearn.metrics import brier_score_loss

import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
cols_for_pred = [
#     'I(ColleyRating_Home - ColleyRating_Away)',
#     'FTEProb_Home',
    'I(FTEImportance_Home - FTEImportance_Away)',
    'I(FTESpi_Home - FTESpi_Away)',
#     'I(FTEProb_Home - FTEProb_Away)',
    
#     'USProb_Home',
#     'I((AvgFBREFORating_Home * AvgFBREFDRating_Away) - (AvgFBREFORating_Away * AvgFBREFDRating_Home))',
#     'I(AvgFBREFORating_Home - AvgFBREFORating_Away)',
#     'I(AvgFBREFDRating_Away - AvgFBREFDRating_Home)',
#     'I(AvgFBREFSCA_Home - AvgFBREFSCA_Away)',
    'I(AvgFBREFORating2_Home - AvgFBREFORating2_Away)',
    'I(AvgFBREFDRating2_Away - AvgFBREFDRating2_Home)',
#     'I((AvgFBREFORating2_Home * AvgFBREFDRating2_Away) - (AvgFBREFORating2_Away * AvgFBREFDRating2_Home) + 0.371)'
    
]

glm_str = "USProb_Home ~ " + '+'.join(cols_for_pred)
print(glm_str)

result_glm = smf.glm(glm_str, 
             data = final_df[final_df['Season'].isin(['2017-2018', '2018-2019'])], 
             family = sm.families.Binomial()
).fit()
result_glm.summary()
```

```python
print("Validation Brier Score: ", brier_score_loss(final_df[final_df['Season'].isin(['2019-2020'])]['Win_Home'], final_df[final_df['Season'].isin(['2019-2020'])]['FTEProb_Home']))
```

```python
print("Validation Brier Score: ", brier_score_loss(final_df[final_df['Season'].isin(['2019-2020'])]['Win_Home'], result_glm.predict(final_df[final_df['Season'].isin(['2019-2020'])])))
```
